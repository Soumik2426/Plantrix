```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using:", device)
print("GPU:", torch.cuda.get_device_name(0))
```

```python
from collections import Counter
import os

# List all class folders
classes = os.listdir(DATA_DIR)
print("Classes found:")
for c in classes:
    print("-", c)

# Count images inside each class
print("\nImage count per class:")
for c in classes:
    folder = os.path.join(DATA_DIR, c)
    count = len([name for name in os.listdir(folder) if os.path.isfile(os.path.join(folder, name))])
    print(f"{c}: {count} images")

# Total image count
total_images = sum(len(os.listdir(os.path.join(DATA_DIR, c))) for c in classes)
print("\nTotal images in dataset:", total_images)
```

```python
IMAGE_SIZE = 224
BATCH_SIZE = 64     # RTX 4090 can handle large batch size
EPOCHS = 20
DATA_DIR = "Dataset"   # CHANGE THIS TO YOUR FOLDER
```

```python
train_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.RandomResizedCrop(IMAGE_SIZE, scale=(0.8, 1.0)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

test_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
```

```python
full_dataset = datasets.ImageFolder(DATA_DIR, transform=train_transform)
class_names = full_dataset.classes
num_classes = len(class_names)

print("Classes:", class_names)
print("Total images:", len(full_dataset))
```

```python
train_size = int(0.7 * len(full_dataset))
val_size = int(0.2 * len(full_dataset))
test_size = len(full_dataset) - train_size - val_size

train_ds, val_ds, test_ds = random_split(full_dataset, [train_size, val_size, test_size])
```

```python
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False)
```

```python
images, labels = next(iter(train_loader))
plt.figure(figsize=(10, 10))
for i in range(9):
    img = images[i].permute(1, 2, 0).numpy()
    img = np.clip(img * np.array([0.229, 0.224, 0.225]) + 
                  np.array([0.485, 0.456, 0.406]), 0, 1)
    plt.subplot(3, 3, i+1)
    plt.imshow(img)
    plt.title(class_names[labels[i]])
    plt.axis("off")
plt.show()
```

```python
model = models.efficientnet_b0(weights="IMAGENET1K_V1")
model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
model = model.to(device)
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005, weight_decay=0.0001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
scaler = torch.amp.GradScaler(device="cuda")
```

```python
for epoch in range(EPOCHS):
    model.train()
    running_loss = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()

    scheduler.step()
    print(f"Epoch {epoch+1}/{EPOCHS} - Loss: {running_loss / len(train_loader):.4f}")
```
